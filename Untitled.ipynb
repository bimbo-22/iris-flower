```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# load iris dataset
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
# split data into X and y for train test split
X = iris.drop(['species'], axis = 1)
y = iris['species']
```

```python
# one hot encode y 
from sklearn.preprocessing import LabelBinarizer
```

```python
encoder = LabelBinarizer()
```

```python
y = encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# using 20% of data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
# scale train and test data
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
# dont fit just transform test
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
# we have 4 flower features 
model.add(Dense(4, activation='relu', input_shape=[4,]))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience = 10)
```

```python
model.fit(x=X_train,y=y_train,epochs=300,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss[['loss', 'val_loss']].plot()
```

```python
loss[['accuracy', 'val_accuracy']].plot()
```

```python
model.evaluate(X_test,y_test, verbose = 0 )
```

```python
# 89% acc is good on to deploy
# retrain on all available data 
```

```python
# use epochs in training
epochs = len(loss)
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
model = Sequential()
# we have 4 flower features 
model.add(Dense(4, activation='relu', input_shape=[4,]))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(scaled_X,y,epochs=epochs)
```

```python
model.save('iris_model.keras')
```

```python
from tensorflow.keras.models import load_model
import joblib
import numpy as np
flower_model = load_model('iris_model.keras')
flower_scaler = joblib.load('iris_scaler.pkl')

# prediction function
def prediction(model,scaler,sample_json):
    sepal_length = sample_json['sepal_length']
    sepal_width = sample_json['sepal_width']
    petal_length = sample_json['petal_length']
    petal_width = sample_json['petal_width']

    # get data into an array to prepare for prediction
    flower = [[sepal_length,sepal_width,petal_length,petal_width]]
    # scale
    flower = scaler.fit_transform(flower)
    # classes
    classes = np.array(['setosa', 'versicolor', 'virginica'])
    # gets the index of predicted class
    result = np.argmax(model.predict(flower), axis=-1)[0]
    # return corresponding class to index
    return classes[result]


    
```

```python
prediction(flower_model,flower_scaler,flower)
```
